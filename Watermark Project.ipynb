```python
# Subsampling
from skimage.transform import rescale, resize, downscale_local_mean
def subsample(im, factor):
    return im[:: factor, :: factor]
im1 = imread('watermark.png') 
im_s= subsample(im1,2)
print(im_s.shape)
plt.figure()
imshow(im_s)
plt.show


# encore mieux: le rescale (more blurry if ord = 1)
im_s2 = rescale (im1, 0.5, order =0)
print(im_s2.shape)
plt.figure()
imshow(im_s2)
plt.show
```
