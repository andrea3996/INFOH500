```python
#Source: https://www.blog.pythonlibrary.org/2017/10/17/how-to-watermark-your-photos-with-python/

from PIL import Image
import matplotlib.pyplot as plt
 

def watermark_picture(input_image_path,output_image_path,watermark_image_path,position):
    base_image = Image.open(input_image_path)
    watermark = Image.open(watermark_image_path)
    # add watermark to your image
    base_image.paste(watermark, position, watermark)
    #base_image.show()
    base_image.save(output_image_path)
    plt.imshow(base_image)

 

im = Image.open('watermark.png')

im = im.convert("RGBA")
im.putalpha(128) #Transparence
datas = im.getdata()


newData = []
for item in datas:
    if item[0] == 0 and item[1] == 0 and item[2] == 0:
        newData.append((0, 0, 0, 0))
    else:
        if item[0] < 150:
            newData.append((0, 0, 0, 0))
        else:
            newData.append(item)
            #print(item)

 

im.putdata(newData)
im.save("Logo_blanc.png", "PNG")

watermark_picture('ex_enhance.jpg', 'final.jpg', "Logo_blanc.png", (0,800))
```
